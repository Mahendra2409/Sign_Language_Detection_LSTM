```python
def extract_hand_keypoints(results):
    # Initialize empty arrays for both hands
    lh = np.zeros(21 * 3)
    rh = np.zeros(21 * 3)
    
    # Check if any hands are detected
    if results.multi_hand_landmarks:
        if len(results.multi_hand_landmarks) > 0:  # Check for left hand landmarks
            lh = np.array([[res.x, res.y, res.z] for res in results.multi_hand_landmarks[0].landmark]).flatten()
        if len(results.multi_hand_landmarks) > 1:  # Check for right hand landmarks
            rh = np.array([[res.x, res.y, res.z] for res in results.multi_hand_landmarks[1].landmark]).flatten()
    
    # Combine left hand and right hand keypoints
    return np.concatenate([lh, rh])
```

## 0. Build and Train Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, Dense
from tensorflow.keras.callbacks import TensorBoard
```

```python
log_dir = os.path.join('Logs_ISL2.O')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
(len(label_map))
```

```python
# model = RandomForestClassifier()
```

```python
model = Sequential()

# Input layer
model.add(Dense(256, activation='relu', input_shape=(126,)))  # First hidden layer
model.add(Dropout(0.3)) 
# Hidden layers
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.3))  # Regularization to prevent overfitting
model.add(Dense(64, activation='relu'))

# Output layer
model.add(Dense(len(label_map), activation='softmax'))  # Output layer with softmax activation
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
model.fit(x_train, y_train,
                    validation_data=(x_test, y_test),
                    epochs=2000,
                    batch_size=64)
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```

```python
def draw(image, results):
    if results.multi_hand_landmarks:  # Check if landmarks are detected
        for hand_landmarks in results.multi_hand_landmarks:
            # Draw landmarks and connections
            mp_drawing.draw_landmarks(
                image,
                hand_landmarks,
                mp_hands.HAND_CONNECTIONS,
                mp_drawing_styles.get_default_hand_landmarks_style(),
                mp_drawing_styles.get_default_hand_connections_style()
            )
    else:
        print("No hands detected")
    return image  # Return the image with drawn landmarks
```
